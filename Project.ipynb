```python
#Aggregate total ammount by region for active custojmers only.

active_customers = customer_df[customer_df["Status"]=="Active"]

active_transactions = pd.merge(active_customers, transactions_df, on="CustomerID")

revenue_by_region = active_transactions.groupby("Region")["Amount"].sum().sort_values(ascending=False)

print(revenue_by_region)
#Visualization

plt.figure(figsize=(10,6))
sns.barplot(x=revenue_by_region.index, y= revenue_by_region.values)

plt.title("Total Revenue by Region (Active Customers)")
plt.xlabel("Region")
plt.ylabel("Total Revenue")
plt.show()        
```

```python
#Plot month-wise revenue trend for 2022

#Step 1: Ensure transaction date is in datetime format.

#Step 2: Filter for year 2024 only.

#Step 3: Extract "Month" as (Year-Month)

#Step 4: Group and sum amount by amount

#Step 5: Plotting

#1.
transactions_df["TransactionDate"] = pd.to_datetime(transactions_df["TransactionDate"], errors="coerce", format="%d-%m-%Y")
transactions_df.head()

#2.
transactions_2024 = transactions_df[transactions_df['TransactionDate'].dt.year==2024]

#3.

transactions_2024["Month"] = transactions_2024["TransactionDate"].dt.to_period("M")

#4.

monthly_revenue = transactions_2024.groupby("Month")["Amount"].sum().sort_index()

#5.

plt.figure(figsize=(10,6))

monthly_revenue.plot(kind="line", marker = "o")
plt.title("Monthly revenue for 2024")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()    
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=monthly_revenue.index, y= monthly_revenue.values)

plt.title("Total Revenue by Region (Active Customers)")
plt.xlabel("Region")
plt.ylabel("Total Revenue")
plt.show() 
```

```python
# Find out the Correlation between subscription type and churn

#Find churn rate for annual vs monthly subscribers.

subscription_churn = pd.merge(subscriptions_df,churn_df, on = "CustomerID", how="left")

# We want to know which subscribed customers have churned by adding churn beside their subscription.

subscription_churn["Churned"] = np.where (subscription_churn["ChurnID"].isna(),"Not Churned", "Churned")

#Create a new column named as "Churned"
#using where function to check: if "ChurnID" is missing (Nan), didn't churn

churn_by_plan = subscription_churn.groupby(["PlanType", "Churned"]).size().unstack().fillna(0)

#size(): counts how many customers are in each group
#unstack(): reshapes the tables, churned values become column headers

plt.figure(figsize=(10,6))
churn_by_plan.plot(kind= "bar", stacked=True)

plt.title("Churn count by subscription type")

plt.ylabel("Number of customers")

plt.show()
```

```python
#Show how churn reasons are distributed across regions. Create a heatmap of churn reasons vs region.

churn_reasons_region = pd.merge(churn_df, customer_df[["CustomerID", "Region"]], on="CustomerID", how="left")

pivot = churn_reasons_region.pivot_table(index="Reason", columns="Region", aggfunc="size", fill_value=0)

#Heat Map

plt.figure(figsize=(10,6))
sns.heatmap(pivot, annot=True)
plt.title("Churn reasons across regions")
plt.xlabel("Region")
plt.ylabel("Churn Reasons")
plt.show()
```

```python
#Calculate the average days of subscription for each plan


#Stes 1: Convert Start date & End date to date time

subscriptions_df["StartDate"]= pd.to_datetime(subscriptions_df["StartDate"], errors="coerce", format="%d-%m-%Y")
subscriptions_df["EndDate"]= pd.to_datetime(subscriptions_df["EndDate"], errors="coerce", format="%d-%m-%Y")

#Steps 2: Calculate the subscriptions duration in days

subscriptions_df["Duration"] = (subscriptions_df["EndDate"] - subscriptions_df["StartDate"]).dt.days

#Steps 3 : Average duration grouped by plan type

avg_duration = subscriptions_df.groupby("PlanType")["Duration"].mean()

#steps 4: Plotting
plt.figure(figsize=(10,6))
avg_duration.plot(kind="bar")
plt.title("Average subscritpion Duration by plan type")
plt.xlabel("Plan Type")
plt.show()
```

```python
#Analyze the customer spending patterns.

spend_per_customer = transactions_df.groupby("CustomerID")["Amount"].sum().reset_index()

#Histogram

#Step 1: Divide customers in group
#Step 2: Find the average for each group
#Step 3: Plot

plt.figure(figsize=(10,6))
sns.histplot(spend_per_customer["Amount"], bins=20, kde=True, color = "coral")
plt.title("Distribution of total spend per customer")
plt.show()
```

```python
#Identify the top 10 spenders across all transactions (bar)

#top_spenders = transactions_df.groupby('CustomerID')['Amount'].sum().sort_values(ascending=False).head(10)

top_customers = spend_per_customer.sort_values(by="Amount", ascending=False).head(10)

top_customers_names = pd.merge(top_customers, customer_df[["CustomerID", "FirstName", "LastName"]], on="CustomerID")

plt.figure(figsize=(10, 6))
sns.barplot(x="Amount", y=top_customers_names["FirstName"]+" "+top_customers_names["LastName"],data = top_customers_names)
plt.title("Top 10 customers by Total Spend")
plt.xlabel("Total Spend")
plt.ylabel("Customer")
plt.show()
```

```python
#Which month has the highest churn behavior (bar chart)

# Make sure the date column is datetime type
churn_df['ChurnDate'] = pd.to_datetime(churn_df['ChurnDate'], dayfirst=True)

# Extract the month and year for grouping
churn_df['churn_month'] = churn_df['ChurnDate'].dt.to_period('M')

# Count churns per month
churn_counts = churn_df['churn_month'].value_counts().sort_index()

# Get the month with highest churn
highest_churn_month = churn_counts.idxmax()
highest_churn_value = churn_counts.max()

print(f"The month with the highest churn is {highest_churn_month} with {highest_churn_value} churns.")
```

```python
plt.figure(figsize=(10,6))
churn_counts.plot(kind='bar', color='coral')
plt.title('Churns per Month')
plt.xlabel('Month')
plt.ylabel('Number of Churns')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
